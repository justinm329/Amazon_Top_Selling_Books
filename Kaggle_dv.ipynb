```python
# get some baisc info from the dataset
df.info()
```

```python
# check for duplicate rows
duplicates = df[df.duplicated("Author")]
duplicates.head()
```

```python
# If we were to expand out we would see multiple authors in that row
# drop the duplicate authors in this dataframe
drop_dup = duplicates.drop_duplicates(subset=["Author"], keep = "first")
drop_dup.tail()
```

```python
# check the number of rows are left after dropping the duplicates
# When we first did the .info above there was 550 entries now there is only 118
drop_dup.info()
```

```python
# check to see if there any null values
df_isnull = drop_dup.isnull().sum()
df_isnull
```

```python
# check the datatypes of the dataframe
drop_dup.dtypes
```

```python
# select certain columns to make a new dataframe to graph
new_df = drop_dup["Genre"].to_frame()
```

```python
# Create new dataframe with genre user rating and book_name
new_df["User_Rating"] = drop_dup["User Rating"]
new_df["Book_name"] = drop_dup["Name"]
new_df["Reviews"] = drop_dup["Reviews"]
new_df.head()
```

```python
# sort the new_df by the genre of the books
sorted_df = new_df.sort_values("Genre")
```

```python
fiction_df = sorted_df[sorted_df['Genre']=="Fiction"].reset_index(drop=True)
fiction_df.head()
```

```python
nonfiction_df = sorted_df[sorted_df['Genre']=="Non Fiction"].reset_index(drop=True)
nonfiction_df.head()
```

```python
top_10_nonfiction = nonfiction_df.sort_values("User_Rating", ascending=False).head(10)
top_10_nonfiction
```

```python
top_10_nonfiction.hvplot.bar(x="Book_name",
                             y="User_Rating",
                            title="Top 10")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
bar = sns.barplot(x=top_10_nonfiction['Book_name'],
           y=top_10_nonfiction['User_Rating'],
           );
bar.set_xticklabels(top_10_nonfiction['Book_name'],rotation=90);
bar.set_title("Top 10");
```
